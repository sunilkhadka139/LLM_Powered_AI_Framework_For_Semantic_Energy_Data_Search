## Crawl and Save Data

```python
# Finl all urls inside BEMcyclopedia
url = 'https://bemcyclopedia.com/wiki/BEMcyclopedia_-_The_Energy_Modeling_Knowledgebase'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
 
list_of_urls = []
for link in soup.find_all('a'):
    list_of_urls.append(link.get('href'))

list_of_urls = ['https://bemcyclopedia.com' + x for x in list_of_urls[1:] if x[0:6] == '/wiki/']
list_of_urls.append('https://bemcyclopedia.com/wiki/Schematic_Design')

list_of_urls = [*set(list_of_urls)]



# aggregate texts from each url
print("Crawl and aggregate texts from each url")
result = {}
for idx, url in enumerate(list_of_urls):
    print(f"Processing: {url}")
    try:
        data = urllib.request.urlopen(url)
        text = []
        for line in data:
            line = line.decode('utf-8')
            text.append(line)

        result[f"text{idx}"] = text
    except:
        print(f'    unable to load: {url}')

        

# Save the crawled text locally
data_path = "data/bemcyclopedia_unprocessed.txt"
with open(data_path, "w", encoding="utf-8") as f: 
    for key, value in result.items(): 
        f.write('%s:%s\n' % (key, value)) 
print(f"\nCrawled data saved to {data_path}")
```

## Postprocess Sentences

```python
# Merge all sentences to one string
def url_to_sentences(url):
    # Fetch webpage
    AttributeErrorresponse = requests.get(url)
    
    html = AttributeErrorresponse.text

    # Parse HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Remove script and style elements
    for script in soup(['script', 'style']):
        script.decompose()

    # Get text
    text = soup.get_text()

    # Split into sentences
    sentences = sent_tokenize(text)
    return sentences

total_sentences = []
for url in list_of_urls:
    total_sentences += url_to_sentences(url)
total_sentences_one_string = ' '.join(total_sentences)



# split the one string to list, each element in the list is a sentence
total_sentences = total_sentences_one_string.split('\n\n')



# Clean sentence: filter out sentence with less than 20 characteristics
total_sentences = [x for x in total_sentences if len(x) > 20]

print(f"total_sentences: {total_sentences}")
```

## Build a Search Index

```python
co = cohere.Client('qYOIHoiJz1P3Jo7BeqdgxApGv0kTiCInhvGHW0MO')
response = co.embed(
    texts=total_sentences,
).embeddings


# Check the dimensions of the embeddings
embeds = np.array(response)

# Create the search index, pass the size of embedding
search_index = AnnoyIndex(embeds.shape[1], 'angular')
# Add all the vectors to the search index
for i in range(len(embeds)):
    search_index.add_item(i, embeds[i])

search_index.build(10) # 10 trees
search_index.save('test.ann')
```

## Searching Relavant Sentences

```python
def search_andrews_article(query):
    # Get the query's embedding
    query_embed = co.embed(texts=[query]).embeddings
    
    # Retrieve the nearest neighbors
    similar_item_ids = search_index.get_nns_by_vector(query_embed[0],
                                                    10,
                                                  include_distances=True)

    search_results = np.array(total_sentences)[similar_item_ids[0]]
    
    return search_results


# Print the top 10 relavent sentences with the question: "what is BEMcyclopedia?"
results = search_andrews_article(
    "what is BEMcyclopedia?"
)

print(results)
```

## Generating Answers

```python
def ask(question, num_generations=1):
    
    # Search the text archive
    results = search_andrews_article(question)

    # Get the top result
    context = results

    # Prepare the prompt
    prompt = f"""
    Excerpt from the BEMcyclopedia: https://bemcyclopedia.com/index.html: 
    {context}
    Question: {question}
    
    Extract the answer of the question from the text provided. 
    If the text doesn't contain the answer, 
    reply that the answer is not available."""
    
#     print(prompt)

    prediction = co.generate(
        prompt=prompt,
        max_tokens=700,
        model="command-nightly",
        temperature=1,
        num_generations=num_generations
    )

    return prediction.generations
```

```python
results = ask(
    "what is the purpose of BEMcyclopedia?",
)

print(results[0])
```
